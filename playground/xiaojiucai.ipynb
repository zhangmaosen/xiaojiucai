```python
import os
import pandas as pd
from torchvision.io import read_image
import torch
from torch.utils.data import Dataset
from torchvision import datasets
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt
import pathlib
from torchvision.transforms import v2

class StockDataset(Dataset):
    def __init__(self, csv_dir, transform=None, target_transform=None):
        self.stock_list = os.listdir(csv_dir)
        self.csv_dir = csv_dir
        self.transform = transform
        self.target_transform = target_transform

    def __len__(self):
        return len(self.stock_list)

    def __getitem__(self, idx):
        stock_price_path = os.path.join(self.csv_dir, self.stock_list[idx])
        stock_name = pathlib.Path(stock_price_path).stem
        colnames = ['date', "movement_percent", "open_price", "high_price", "lowprice", "close_price", "volume"]
        df = pd.read_csv(stock_price_path,sep='\t',header=0, names=colnames)
        prices = df.iloc[:,1:].values
        prices_t = torch.tensor(prices)
        prices_t = prices_t.permute(1,0).unsqueeze(2)
        prices_t = v2.Resize([1024,1])(prices_t)
        # if self.transform:
        #     image = self.transform(image)
        # if self.target_transform:
        #     label = self.target_transform(label)
        return prices_t, stock_name
```
